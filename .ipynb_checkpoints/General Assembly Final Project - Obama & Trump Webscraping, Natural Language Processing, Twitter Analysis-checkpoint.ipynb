### Webscraping & Data Cleaning ###

```python
obama_first_url = 'https://en.wikisource.org/wiki/Barack_Obama%27s_First_State_of_the_Union_Address'
obama_second_url = 'https://en.wikisource.org/wiki/Barack_Obama%27s_Second_State_of_the_Union_Address'

trump_first_url = 'https://en.wikisource.org/wiki/Donald_Trump%27s_Address_to_a_Joint_Session_of_Congress'
trump_second_url = 'https://en.wikisource.org/wiki/Donald_Trump%27s_State_of_the_Union_Address_2018'
```

```python
first_response = requests.get(obama_first_url)
second_response = requests.get(obama_second_url)

third_response = requests.get(trump_first_url)
fourth_response = requests.get(trump_second_url)
```

```python
osoup = BeautifulSoup(first_response.text, 'html.parser')
osoup2 = BeautifulSoup(second_response.text, 'html.parser')

tsoup1 = BeautifulSoup(third_response.text, 'html.parser')
tsoup2 = BeautifulSoup(fourth_response.text, 'html.parser')

```

```python
all_p = osoup.findAll('p')
all_p2 = osoup2.findAll('p')
all_p3 = tsoup1.findAll('p')
all_p4 = tsoup2.findAll('p')

```

```python
all_p = all_p[1:len(all_p)-1]
all_p2 = all_p2[1:len(all_p2)-1]
all_p3 = all_p3[1:len(all_p3)-1]
all_p4 = all_p4[1:len(all_p4)-1]
```

```python
ospeech1_list = []
ospeech2_list = []
tspeech1_list = []
tspeech2_list = []

for item in all_p:
    ospeech1_list.append(item.text.strip())
    
for item in all_p2:
    ospeech2_list.append(item.text.strip())
    
for item in all_p3:
    tspeech1_list.append(item.text.strip())
    
for item in all_p4:
    tspeech2_list.append(item.text.strip())
    
```

```python
ospeech1_string = ""
for item in ospeech1_list:
    ospeech1_string = ospeech1_string + str(item) + " "
    
ospeech2_string = ""
for item in ospeech2_list:
    ospeech2_string = ospeech2_string + str(item) + " "
    
tspeech1_string = ""
for item in tspeech1_list:
    tspeech1_string = tspeech1_string + str(item) + " "
    
tspeech2_string = ""
for item in tspeech2_list:
    tspeech2_string = tspeech2_string + str(item) + " "
```

```python
from nltk import sent_tokenize
# Converts giant string into sentences based on punctuation and whitespace
sentences_ospeech1 = sent_tokenize(ospeech1_string)
sentences_ospeech2 = sent_tokenize(ospeech2_string)
sentences_tspeech1 = sent_tokenize(tspeech1_string)
sentences_tspeech2 = sent_tokenize(tspeech2_string)

from nltk.tokenize import word_tokenize
# Splits strings into "tokens" (words) for analysis
tokens_o1 = word_tokenize(ospeech1_string)
tokens_o2 = word_tokenize(ospeech2_string)
tokens_t1 = word_tokenize(tspeech1_string)
tokens_t2 = word_tokenize(tspeech2_string)

# # Lowers everything to filter out stop words later (Not lowering due to Named Entity Recognition later)
# tokens_o1 = [w.lower() for w in tokens_o1]
# tokens_o2 = [w.lower() for w in tokens_o2]
# tokens_t1 = [w.lower() for w in tokens_t1]
# tokens_t2 = [w.lower() for w in tokens_t2]

import string
# Uses string.punctuation & maketrans to strip out punctuation characters from the tokens list

table = str.maketrans('', '', string.punctuation)
stripped_o1 = [w.translate(table) for w in tokens_o1]
stripped_o2 = [w.translate(table) for w in tokens_o2]
stripped_t1 = [w.translate(table) for w in tokens_t1]
stripped_t2 = [w.translate(table) for w in tokens_t2]

# Removes all non-alphabet characters
words_o1 = [word for word in stripped_o1 if word.isalpha()]
words_o2 = [word for word in stripped_o2 if word.isalpha()]
words_t1 = [word for word in stripped_t1 if word.isalpha()]
words_t2 = [word for word in stripped_t2 if word.isalpha()]

# Creates stop words variable in English to remove stop words
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))

clean_words_o1 = [w for w in words_o1 if not w in stop_words]
clean_words_o2 = [w for w in words_o2 if not w in stop_words]
clean_words_t1 = [w for w in words_t1 if not w in stop_words]
clean_words_t2 = [w for w in words_t2 if not w in stop_words]
```

### Twitter Analysis ### 

```python
consumer_key = 'NA'
consumer_secret = 'NA'
access_token = 'NA'
access_token_secret = 'NA'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
obama_twitter = api.get_user('POTUS44')
trump_twitter = api.get_user('realDonaldTrump')
```

```python
obama_tweets = []
for tweet in obama_twitter.timeline(count=(200), since_id = 600324682190053375, max_id = 822446982648201217):
    obama_tweets.append(tweet.text)
    
trump_tweets = []
for tweet in trump_twitter.timeline(count=(200), since_id = 796315640307060737, max_id = 1031689704067018752):
    trump_tweets.append(tweet.text)
    
```

```python
obama_tweets_string = ""
for item in obama_tweets:
    obama_tweets_string = obama_tweets_string + str(item) + " "
    
trump_tweets_string = ""
for item in trump_tweets:
    trump_tweets_string = trump_tweets_string + str(item) + " "
    
```

#### Verify Twitter Results ####

```python
for tweet in range(0,3):
    print(obama_tweets[tweet] + "\n")
print("")
for tweet in range(0,3):
    print(trump_tweets[tweet] + "\n")
```

### Create Markov Chain Generator AKA Markovify in Python ###

```python
obama_tweet_model = markovify.Text(obama_tweets_string)
trump_tweet_model = markovify.Text(trump_tweets_string)

# Prints five generated sentences of 140 characters or less

for i in range(5):
    print("Obama: " + obama_tweet_model.make_short_sentence(140) + "\n")
    
for i in range(5):
    print("Trump: " + trump_tweet_model.make_short_sentence(140) + "\n")

```

```python
obama_SOTU_model_1 = markovify.Text(ospeech1_string)
obama_SOTU_model_2 = markovify.Text(ospeech2_string)
trump_SOTU_model_1 = markovify.Text(tspeech1_string)
trump_SOTU_model_2 = markovify.Text(tspeech2_string)

# Prints five generated sentences of based on SOTU addresses

for i in range(5):
    print("Obama: " + obama_SOTU_model_1.make_sentence() + "\n")
    
    
for i in range(5):
    print("Trump: " + trump_SOTU_model_1.make_sentence() + "\n")

```

```python
# Clean up tweet strings

obama_tweets_string
trump_tweets_string

# Removes all non-alphabet characters & http links

obama_tweets_clean = re.sub(r'^A-Za-z0-9','',obama_tweets_string)
obama_tweets_clean = re.sub(r"http\S+",'',obama_tweets_string)

trump_tweets_clean = re.sub(r'^A-Za-z0-9','',trump_tweets_string)
trump_tweets_clean = re.sub(r"http\S+",'',trump_tweets_string)

# Creates stop words variable in English to remove stop words

stop_words = set(stopwords.words('english'))

obama_tweets_words = word_tokenize(obama_tweets_clean)
trump_tweets_words = word_tokenize(trump_tweets_clean)

obama_tweets_clean_no_stopwords = [w for w in obama_tweets_words if not w in stop_words]
trump_tweets_clean_no_stopwords = [w for w in trump_tweets_words if not w in stop_words]
```

### Sentiment Analysis: SOTU & Twitter ###

* Subjectivity - Addresses how objective or subjective a word / statement is from 0 to 1
* Polarity - Addresses how negative or positive a a word / statement is is from -1 to 1

```python
# Using our NTLK sentences above which we generated via the sentence tokenizer, 
# we can quickly calculate average subjectivity & polarity from their SOTU addresses

o1_blob = TextBlob(ospeech1_string)
o2_blob = TextBlob(ospeech2_string)
t1_blob = TextBlob(tspeech1_string)
t2_blob = TextBlob(tspeech2_string)
```

```python
o1_subj = []
o1_pol = []

for sentence in o1_blob:
    o1_subj.append(o1_blob.sentiment.subjectivity)
    o1_pol.append(o1_blob.sentiment.polarity)
    
o1_subj_mean = np.mean(o1_subj)
o1_pol_mean = np.mean(o1_pol)

print(o1_subj_mean, o1_pol_mean)
```

```python
o2_subj = []
o2_pol = []

for sentence in o2_blob:
    o2_subj.append(o2_blob.sentiment.subjectivity)
    o2_pol.append(o2_blob.sentiment.polarity)
    
o2_subj_mean = np.mean(o2_subj)
o2_pol_mean = np.mean(o2_pol)

print(o2_subj_mean, o2_pol_mean)
```
